```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])
```

```python
print("added : \n", x + y)
print("Substracted : \n", v - w)
print("Multiplied by 3 : \n", w * 3)
print("Squared : \n", x ** 2)
```

```python
print("dot product of v and w : ", np.dot(v,w))
print("dot product of x and y : ", np.dot(x,y))
print("dot product of x and v : ", np.dot(x,v))
```

```python
import numpy as np

# Given arrays
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])
```

```python
# Compute inverse of A
A_inv = np.linalg.inv(A)

# Multiply A by its inverse
identity_check = np.dot(A, A_inv)

print("A * A^-1 =\n", identity_check)

# Check if approximately identity
print("Is approximately identity?\n", np.allclose(identity_check, np.eye(2)))
```

```python
import numpy as np

# Coefficient matrix A
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant matrix B
B = np.array([-1, -3, 9])

# Compute inverse of A
A_inv = np.linalg.inv(A)

# Solve for X
X = np.dot(A_inv, B)

print("Solution [x, y, z] =", X)
```
