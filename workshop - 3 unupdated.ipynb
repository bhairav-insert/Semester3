```python
# Step 0: Handle missing Age values by filling with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Step 1: Add fare_per_year column
df['fare_per_year'] = df['Fare'] / df['Age']

# Step 2: Subset rows where fare_per_year > 5
high_fare_age = df[df['fare_per_year'] > 5]

# Step 3: Sort by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# Step 4: Select only Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

# Step 5: Look at the result
print(result.head(10))  # top 10 passengers
```

```python
# Step 0: Handle missing Age values
df['Age'].fillna(df['Age'].median(), inplace=True)

# Step 1: Add fare_per_class column
df['fare_per_class'] = df['Fare'] / df['Pclass']

# Step 2: Subset adult males (Age >= 18 and Sex == 'male')
adult_males = df[(df['Age'] >= 18) & (df['Sex'] == 'male')]

# Step 3: Sort adult_males by descending fare_per_class
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# Step 4: Select only Name, Age, and fare_per_class columns
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# Step 5: Look at the result
print(result.head(10))  # top 10 adult males by fare per class
```

```python
import numpy as np

# Basic sanity checks: ensure key columns exist
required_cols = {"Pclass", "Fare", "Age"}
missing = required_cols - set(df.columns)
if missing:
    raise ValueError(f"Your dataset is missing required columns: {missing}. "
                     f"Expected columns include at least: {required_cols}")

# Optional: show brief info (comment out if not needed)
# print(df[["Pclass", "Fare", "Age"]].head())

# === Question 1: Percent of total fare revenue from each passenger class ===
# Steps:
# 1. Total fare across all passengers
total_fare = df["Fare"].sum(skipna=True)

# 23. Total fare by class (1, 2, 3) using groupby
fare_by_class = df.groupby("Pclass", dropna=False)["Fare"].sum()

# 45. Proportion (percentage) per class of the overall fare
fare_pct_by_class = (fare_by_class / total_fare) * 100

# Make result tidy and sorted by class
fare_pct_by_class = fare_pct_by_class.sort_index()

# === Question 2: Percent of passengers by age group (child, adult, senior) ===
# Define age groups:
# - child: age < 18
# - adult: 18 <= age < 65
# - senior: age >= 65
# We will exclude rows with missing Age from the age-group calculation, then
# report the share relative to all passengers and also relative to passengers with known age.

# Create age_group with pd.cut; first dropna for Age grouping
age_bins = [-np.inf, 18, 65, np.inf]
age_labels = ["child", "adult", "senior"]

df_age_known = df.dropna(subset=["Age"]).copy()
df_age_known["age_group"] = pd.cut(df_age_known["Age"], bins=age_bins, labels=age_labels, right=False)

# 2. Total passengers (overall)
total_passengers = len(df)

# Also compute total with known age
total_with_age = len(df_age_known)

# 3. Count passengers in each age group (based on known ages)
age_group_counts_known = df_age_known["age_group"].value_counts().reindex(age_labels).fillna(0)

# 4. Proportions:
#    a) As a percent of ALL passengers (including those with missing ages)
age_group_pct_overall = (age_group_counts_known / total_passengers) * 100

#    b) As a percent of ONLY passengers with known ages
age_group_pct_known = (age_group_counts_known / total_with_age) * 100 if total_with_age > 0 else pd.Series([np.nan]*3, index=age_labels)

# === Display nicely ===
def fmt_pct(series):
    return series.map(lambda x: f"{x:.2f}%")

print("\n=== Question 1: Percent of total fare revenue by passenger class ===")
print("Total fare (all passengers):", f"{total_fare:.2f}")
print("Fare revenue by class:")
print(fare_by_class.map(lambda x: f"{x:.2f}"))
print("\nPercent of total fare revenue by class:")
print(fmt_pct(fare_pct_by_class))

print("\n=== Question 2: Percent of passengers by age group ===")
print(f"Total passengers (overall): {total_passengers}")
print(f"Passengers with known age: {total_with_age}")
print("\nCounts by age group (known ages only):")
print(age_group_counts_known.astype(int))

print("\nPercent of ALL passengers by age group (missing ages counted as 'unknown'):")
print(fmt_pct(age_group_pct_overall))

print("\nPercent of passengers WITH KNOWN AGE by age group:")
print(fmt_pct(age_group_pct_known))

# Optional: also produce a compact DataFrame summary for each question
summary_q1 = (
    pd.DataFrame({
        "Total_Fare_by_Class": fare_by_class,
        "Percent_of_Total_Fare": fare_pct_by_class
    })
    .rename_axis("Pclass")
)

summary_q2_overall = (
    pd.DataFrame({
        "Count (known ages)": age_group_counts_known,
        "Percent of ALL passengers": age_group_pct_overall,
        "Percent of passengers WITH KNOWN AGE": age_group_pct_known
    })
    .rename_axis("age_group")
)

print("\n--- Summary (Q1) ---")
print(summary_q1)

print("\n--- Summary (Q2) ---")
print(summary_q2_overall)
```
